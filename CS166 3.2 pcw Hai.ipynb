```python
#Exercise 11.3

import matplotlib
from pylab import *
%matplotlib inline


n = 100 # size of space: n x n
p = 0.5 # probability of initially alive individuals

def initialize():
    global config, nextconfig, list_density
    config = zeros([n, n])
    for x in range(n):
        for y in range(n):
            config[x, y] = 1 if random() < p else 0
    nextconfig = zeros([n, n])
    list_density = []
    
def observe():
    global config, nextconfig, list_density
    cla()
    list_density.append(sum(config))
    imshow(config, vmin = 0, vmax = 1, cmap = cm.binary)

def update():
    global config, nextconfig, list_density
    for x in range(n):
        for y in range(n):
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    count += config[(x + dx) % n, (y + dy) % n] 
            if count >= 4 or count <=1: #Game of Life rules
                nextconfig[x, y] = 0 
            elif count == 3:
                nextconfig[x, y] = 1
            else:
                pass
    config, nextconfig = nextconfig, config

initialize()
for i in range(200):
    update()
    observe()
```

```python
import numpy as np
x_axis = np.arange(0,200,1)
matplotlib.pyplot.plot(x_axis, list_density)
matplotlib.pyplot.title("Density at each time step")
matplotlib.pyplot.xlabel("Time step")
matplotlib.pyplot.ylabel("Density")
matplotlib.pyplot.show()
```
